```python
with open("./data.data", "rb") as f:
    data = pk.load(f)
```

```python
# 날짜, 삼성전자, 카카오, 엘지전자, 현대차
date, sec, kakao, lge, hdm = data
```

```python
plt.figure(figsize=(20, 10))
plt.rcParams.update({"font.size": 20})
plt.plot(date, sec, label="삼성전자")
plt.plot(date, kakao, label="카카오")
plt.plot(date, lge, label="엘지전자")
plt.plot(date, hdm, label="현대차")
plt.legend()  # 범례
```

```python
plt.figure(figsize=(20, 10))
plt.rcParams.update({"font.size": 20})
plt.plot(date, sec, "o-", label="삼성전자")
plt.plot(date, kakao, "o:", label="카카오")
plt.plot(date, lge, "o--", label="엘지전자")
plt.plot(date, hdm, "o-.", label="현대차")
plt.legend()  # 범례
```

```python
import numpy as np

sec_rate = np.array(sec) / sec[0] - 1
lge_rate = np.array(lge) / lge[0] - 1
kakao_rate = np.array(kakao) / kakao[0] - 1
hdm_rate = np.array(hdm) / hdm[0] - 1
```

```python
plt.figure(figsize=(20, 10))
plt.rcParams.update({"font.size": 20})
plt.plot(date, sec_rate, "o-", label="삼성전자")
plt.plot(date, kakao_rate, "o:", label="카카오")
plt.plot(date, lge_rate, "o--", label="엘지전자")
plt.plot(date, hdm_rate, "o-.", label="현대차")
plt.legend()  # 범례
```

```python
# 가로선 추가

yticks = np.arange(-0.25, 1.75, 0.25)  # <-

plt.figure(figsize=(20, 10))
plt.rcParams.update({"font.size": 20})
plt.axhline(y=0, c="lightgray", zorder=0)  # <-
plt.yticks(ticks=yticks, labels=[f"{y * 100:.0f}%" for y in yticks])  # <-
plt.ylim(-0.5, 1.75)  # <-

plt.plot(date, sec_rate, "o-", label="삼성전자")
plt.plot(date, kakao_rate, "o:", label="카카오")
plt.plot(date, lge_rate, "o--", label="엘지전자")
plt.plot(date, hdm_rate, "o-.", label="현대차")
plt.legend()  # 범례
```
