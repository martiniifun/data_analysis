```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.distplot(df["tip"])
```

```python
sns.displot(df["tip"], kde=True, bins=10)
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"])
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], kind="hex")
```

```python
sns.jointplot(x=df["total_bill"], y=df["tip"], kind="reg")
```

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df, hue="smoker")
```

```python
sns.barplot(x="smoker", y="total_bill", data=df)
```

```python
sns.countplot(x="sex", data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df, palette="Set2")
```

```python
sns.boxplot(x="day", y="total_bill", data=df, palette="Set2", hue="smoker")
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue="smoker", split=True)
```
